```python
team1="Royal Challengers Bangalore"
team2="Chennai Super Kings"
match_count=0
winner_list=[]
win_by_list=[]
for i,j,k,l,m in zip(data["team1"],data["team2"],data["winner"],data["win_by_runs"],data["win_by_wickets"]):
    if i==team1 and j==team2:
        match_count+=1
        winner_list.append(k)
        
        if l!=0:
            msg=str(k)+" won by "+str(l)+" runs"
            win_by_list.append(msg)
        
        if m!=0:
            msg=str(k)+" won by "+str(m)+" wickets"
            win_by_list.append(msg)
            
            
        
       
```

```python
msg1="Total number of matches played between "+str(team1)+"and "+str(team2)+"are:"+str(match_count)
msg2=""
for i in range(match_count):
    msg2=msg2+"Match "+ str(i+1)+ " Result: "+win_by_list[i]+" \n"
```
